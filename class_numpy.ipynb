# 1. 넘파일 다차원 배열(ndarray) 객체 생성
- import numpy as np
- 리스트을 이용하여 다차원 배열 생성

# 2. 넘파이 다차원 배열 간 사칙 연산
  - 피연산자 다차원 배열들의 형상이 같다는 가정
  - +, -, *, / 연산 수행
  - 배열 프로그래밍
  - 선형대수의 벡터화 연산
  - 벡터을 병렬 연산 수행

```python
# 1차원 배열 사칙연산
# 1차원 넘파이 배열 생성
store_a = [1000, 2000, 1500]
store_b = [2000, 1000, 1000]
arr_a = np.array(store_a)
arr_b = np.array(store_b)
# 형태가 같은 1차원 배열 간 연산 : (3,)
print(arr_a + arr_b)
print(arr_a - arr_b)
print(arr_a * arr_b)
print(arr_a / arr_b)
```

# 6. 다차원 배열의 지정된 축 방향으로 원소들을 뒤집기
  - np.flip(배열객체, axis=)

```python
# (4, 5) 구조의 이미지 데이터를 생성 => 상하, 좌우 반전시켜 새로운 이미지 만들기
# 데이터 생성
img = np.arange(20).reshape((4, -1))
print(img)

# (1) 좌우 반전 (axis=1)
h_flip = np.flip(img, axis = 1)
print(h_flip)

# (2) 상하 반전 (axis=0)
v_flip = np.flip(img, axis=0)
print(h_flip)

# (3) 전체 반전 (상하좌우 반전)
all_flip = np.flip(img)
print(all_flip)

imgt = np.swapaxes(img, 0, 1)
print(imgt)
```

# 8. 넘파이 다차원 배열의 인덱싱을 이용한 슬라이싱
- 다차원 배열에서 작은 배열을 추출
- 인덱싱을 이용하여 슬라이싱 범위 지정
- 두 축을 이용해서 특정한 행과 열만 별도로 적용하여 추출함
  - axis 0 ,  axis 1
  - arr[] ; axis 0 사용
  - arr[:, :] ;  두 개의 축을 동시에 사용할 때는 쉼표로 구분
  - arr[: :step] ; axis 0을 이용하여 간격을 두고 추출
  - arr[::step, ::step]; 두 개의 축을 이용하여 간격을 두고 추출


```python
import numpy as np

# ex: 3차원 배열 슬라이싱

# 데이터 생성: 60개 원소로 구성된 (3, 4, 5) 형태의 3차원 배열 생성
tensor = np.arange(60).reshape((3, 4, 5))
print(tensor)
print("="*30)

# (1) 특정 층 전체 가져오기 (첫번째 층)
print(tensor[0, :, :]) # tensor[0] 동일
print("="*30)

# (2) 모든 층에서 특정 '행'만 가져오기 (모든 층의 첫번째 행)
print(tensor[:, 0, :]) # shape(3, 5)
print("="*30)

# (3) 모든 층에서 첫번째 column만
print(tensor[:, :, 0])
print("="*30)

# (4) 정가운데 부분만 즉, 큐브형태로 도려내기
print(tensor[:, 1:3, 1:4])
print("="*30)
```

# 9.  넘파이 다차원 배열의 슬라이싱을 이용한 논리인덱싱
- 특정한 조건을 만족하는 배열의 원소값만을 추출하여 1차원 배열 객체로 반환
- 브로드캐스팅하여 병렬 연산 처리


1. 주어진 배열을 이용하여 조건식을 만족하는 동일한 형상을 가지는 불리언 배열을 생성

2.   주어진 배열에 불리언 배열을 인덱스로 지정


- cf) np.where(조건식, x, y)
  - 주어진 배열의 구조는 유지하면서 조건식에 맞는 원소를 추출하는 기법
  - if ~ else 처럼 동작
  - x는 조건식이 True인 경우 처리
  - y는 False인 경우 처리



```python
import numpy as np

# 2차원 배열 데이터 생성
arr = np.arange(100).reshape(5, -1)
print(arr, "\n")

# ex1: 주어진 배열에서 짝수들만 추출하여 1차원 배열로 생성

# (1) 불리언 배열
arr_bool = arr % 2 == 0 # (5, 20)
print(arr_bool)

# (2) boolean indexing
arr_even = arr[arr_bool]
print(arr_even) # 1차원 배열

# ex2: 주어진 배열에서 20보다 큰 요소들만 추출하여 1차원 배열로 생성
arr_greaterthan_20 = arr[arr > 20]
print(arr_greaterthan_20)

# ex3: 주어진 배열에서 20보다 큰 요소들만 추철
arr_filter = np.where(arr > 20, arr, 0)
```

# 13.난수을 이용한 다차원 배열 만들기

  - randn(개수) - 표준정규분포의 난수 표본 추출
  - randint(최소값,최대값,개수)- 주어진 범위내의 정수 난수 생성. 최대값은 미포함

  - permutation() - 배열을 임의의 순서로 섞음


```python
# 주가 수익률 그래프 시뮬레이션 (평균 0.001, 표준편차 0.02인 난수 1099개)
# 의미: 이런 문제의 불확실성을 난수로 생성하여 향수 1000일 간의 주가를 예측해 보기
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import warnings # 경고 메시지 숨기기
import os

# 1. 환경 설정( 경고 차단 및 한글 폰트 )
warnings.filterwarnings('ignore')
font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
if not os.path.exists(font_path):
    !sudo apt-get install -y fonts-nanum > /dev/null
font_prop = fm.FontProperties(fname=font_path)

# 2. 데이터 준비
np.random.seed(42) # 재현성 보장
daily_values = np.random.normal(0.001, 0.02, 1000)
price = np.cumprod(1 + daily_values) # price.max(), price.min() 이용
print(price)

# 2. simulation : line graph
plt.figure(figsize = (14, 5))
plt.plot(np.arange(1000), price, label = "주가 수익률", color = 'r')
plt.title("주가 수익률 시뮬레이션", fontproperties = font_prop, fontsize = 15)
plt.xlabel("일수(days)", fontproperties = font_prop)
plt.ylabel("가격(price)", fontproperties = font_prop)
plt.xticks(np.arange(0, 1001, 25))
plt.yticks(np.arange(0, 3.7, 0.5))
plt.legend(prop = font_prop)
plt.show()
```

# 14. 다차원 배열 프로그래밍
- 벡터화 연산 (데이터병렬 처리)
  - 배열 간 연산
  - 리덕션
    - np.sum(arr, axis=)
    - np.mean(arr, axis=)
    - arr.sum(axis=0)
    - arr.sum(axis=1)
    - arr.sum(axis=(0,1))
  - 조건 리덕션 원리 (Boolean to Integer)
      - NumPy에서 리덕션이 강력한 이유는 **논리값(True/False)**을 연산하는 순간 **숫자(1/0)**로 자동 변환.

      - np.sum(condition) : 조건이 참(True)인 개수를 구함

      - np.mean(condition) : 조건이 참(True)인 비율을 구함 (예: 0.7이면 70%가 참)
      - 이 리덕션 기법들을 활용해서 실제 데이터셋에서 "이상치(Outlier)"를 찾기

```python
## Ex: 3명의 국어, 영어, 수학 점수에 대한 데이터를 난수로 생성해서
#     원본 데이터, 과목 점수 평균, 학생 점수의 평균에 대한 세 종류의 그래프 분석

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import warnings # 경고 메시지 숨기기
import os

warnings.filterwarnings('ignore')
font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
if not os.path.exists(font_path):
    !sudo apt-get install -y fonts-nanum > /dev/null
font_prop = fm.FontProperties(fname=font_path)

# 데이터 준비
np.random.seed(42)
scores = np.random.randint(60, 101, 12).reshape(3, -1) # (3, 4)
subjects = ['국어', '영어', '수학', '통계']
students = ['영희', '철수', '신한']

# 리덕션 연산: 평균
subjects_avg = np.mean(scores, axis = 0).round(2) # 과목별 평균
student_avg = np.mean(scores, axis = 1).round(2) # 학생별 평균

# 3. 복수 개수 그래프로 시각화
plt.figure(figsize=(16, 6))
plt.suptitle("원본 데이터 및 리덕션 데이터 분석", fontproperties=font_prop, fontsize = 20)
plt.show()
# (1) 원본 데이터 열지도
plt.subplot(1, 3, 1) # 그래프의 위치
im = plt.imshow(scores, cmap = "YlGnBu", aspect="auto")
plt.title("원본 성적 분포", fontproperties = font_prop, fontsize=15)
plt.xlabel('과목명', fontproperties=font_prop)
plt.ylabel('학생명', fontproperties=font_prop)
plt.xticks(np.arange(len(subjects)), subjects, fontproperties=font_prop)
plt.yticks(np.arange(len(students)), students, fontproperties=font_prop)
cbar = plt.colorbar(im)
cbar.set_label("점수")
# (1-1) 특정 좌표에 점수 텍스트를 표기: plt.text(x, y, str())
for i in range(3): # 행 (y축)
  for j in range(4): # 열 (x축)
    plt.text(j, i,  scores[i, j], ha = 'center', va = 'center', color="black", fontsize=15)
# plt.show()


# (2) 과목별 평균 시각화: 막대그래프
plt.subplot(1, 3, 2) # 그래프의 위치
plt.bar(subjects, subjects_avg, color = ['red', 'green', 'blue', 'yellow'], edgecolor='black')
plt.title("과목별 평균 분포", fontproperties = font_prop, fontsize=15)
plt.xlabel('과목명', fontproperties=font_prop)
plt.ylabel('평균점수', fontproperties=font_prop)
plt.xticks(np.arange(len(subjects)), subjects, fontproperties=font_prop)
for i, score in enumerate(subjects_avg):
  plt.text(i, score, str(score), ha="center", va="bottom", color='black', fontsize='12')

# (3) 학생별 평균 시각화: 수평막대 그래프
plt.subplot(1, 3, 3) # 그래프의 위치
plt.barh(students, student_avg, color='lightgreen', edgecolor='black')
plt.title("학생별 평균 분포", fontproperties = font_prop, fontsize=15)
plt.xlabel('평균점수', fontproperties=font_prop)
plt.ylabel('학생명', fontproperties=font_prop)
plt.yticks(fontproperties=font_prop)
for i, score in enumerate(student_avg):
  plt.text(score+1, i, str(score), ha="center", va="bottom", color='black', fontsize='12')
plt.tight_layout()


plt.show()
```

# 15. 넘파일 배열의 결합
- np.concatenate(튜플)
- np.vstack(튜플 또는 리스트)
- np.hstack(튜플 또는 리스트)
- np.r_[ 콤마 사용하여 나열 ] 중간고사 제외
- np.c_[콤마 사용하여 나열] 중간고사 제외
- np.column_stack(리스트)


```python
# Ex1 1차원 배열간 결함
import numpy as np
a = np.array([1, 2, 3]) # (3,)
b = np.array([4, 5, 6]) # (3,)
```

```python
# 세로 쌓기 (수직방향) -> 행 추가
res_stack = np.vstack((a, b))
print(res_stack) # (2, 3)
```

```python

# 가로로 붙이기 (수평방향) -> 열 추가해서 1차원 배열
res_hstack = np.hstack((a, b))
print(res_hstack)
res_concat = np.concatenate((a, b))
print(res_concat)
```

```python

# 주의: column_stack() -> 2차원 배열
res_col = np.column_stack((a, b))
print(res_col)
```

```python

# Ex2 형태가 다른 배열간 결합
arr_1 = np.arange(1, 4) # (3,)
arr_2 = np.arange(4, 10).reshape(2, 3) # (2, 3)
```

```python

# 방법 1
# (1) arr_1 -> (1, 3)
arr_1_2d = arr_1.reshape(-1, 3)
print(arr_1_2d)
combined_2d = np.concatenate((arr_1_2d, arr_2), axis=0)
print(combined_2d)
```

```python

# 방법 2
v_combined = np.vstack((arr_1, arr_2))
print(v_combined)
```
